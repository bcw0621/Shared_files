```python
from scipy import stats
```

```python
#Define the 2 distributions
Black = dfDT['Black']
Total = dfDT['Total']

#Standard Deviation
var_Black = Black.var(ddof=1)
var_Total = Total.var(ddof=1)

sdev = np.sqrt((var_Black +var_Total)/2)

#Calculate the t-statistic
t = (Black.mean()-Total.mean())/(sdev*np.sqrt(2/37))

#Compare with critical t-value
#Degrees of Freedom
DoF = 2*37-2

#p-value after comparison with the t
p = 1-stats.t.cdf(t,df=DoF)

print('t = ' + str(t))
print('p = ' + str(2*p))

t2, p2 = stats.ttest_ind(Black,Total)
print('t = ' + str(t2))
print('p = ' + str(p2))
```
